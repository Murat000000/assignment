```python
import pandas as pd

data = pd.read_csv("worldbank-lifeexpectancy-raw.csv")
data.head()
```

# Dropping missing values and changing string values to float

```python
import numpy as np

# cahnging ".." to missing values
data = data.replace("..", np.nan)

#dropping missing values
data = data.dropna()

data.head()
```

```python
# converting to float
data["GDP per capita, PPP (constant 2011 international $) [NY.GDP.PCAP.PP.KD]"] = data["GDP per capita, PPP (constant 2011 international $) [NY.GDP.PCAP.PP.KD]"].astype(float)
data["Life expectancy at birth, total (years) [SP.DYN.LE00.IN]"] = data["Life expectancy at birth, total (years) [SP.DYN.LE00.IN]"].astype(float)

data.describe()
```

# Making scatter plot and analyzing the data

```python
!pip install matplotlib seaborn

import matplotlib.pyplot as plt
import seaborn as sns

# Scatter plot with Lowess line
plt.figure(figsize=(10, 6))
sns.scatterplot(y="GDP per capita, PPP (constant 2011 international $) [NY.GDP.PCAP.PP.KD]", x="Life expectancy at birth, total (years) [SP.DYN.LE00.IN]", data=data, color = "red", label = "Data points")
sns.regplot(y="GDP per capita, PPP (constant 2011 international $) [NY.GDP.PCAP.PP.KD]", x="Life expectancy at birth, total (years) [SP.DYN.LE00.IN]", data=data, scatter=False, lowess=True, color="blue", label="Trend line")
```

# Summary
As we see from this scatter plot with lowess there is a connection between gdp per capita and life expectancy. In countries with higher gdp per capita, life expectancy tends to be higher and we can guess, anticipate higher life expectancy given higher gdp per capita. However as we see from our data there is a lot of variatiane as gdp per capita goes low, meaning at low gdp per capita data varies widely and we can't say that low life expectancy is due to low gdp per capita results in country, as other factors may contribute significantly to these variations.
